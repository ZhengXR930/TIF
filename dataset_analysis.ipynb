```python
import os
import pandas as pd
import numpy as np
import pickle

dataset_folder = "/scratch_NOT_BACKED_UP/NOT_BACKED_UP/xinran/dataset/combine_drebin"
years = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

dfs_train = []
dfs_test = []
for month in months:
    file_path = os.path.join(dataset_folder, f'2014-{month}', 'features.pkl')
    with open(file_path, 'rb') as f:
        data = pickle.load(f)
    dfs_train.append(data)

for year in years:
    for month in months:
        file_path = os.path.join(dataset_folder, f'{year}-{month}', 'features.pkl')
        with open(file_path, 'rb') as f:
            data = pickle.load(f)
        dfs_test.append(data)

df_train = pd.concat(dfs_train)
print(df_train.shape)

df_test = pd.concat(dfs_test)
print(df_test.shape)

print(df_train.columns)
print(df_test.columns)

```

```python
import os
import pickle
import pandas as pd

dataset_folder = "/scratch_NOT_BACKED_UP/NOT_BACKED_UP/xinran/dataset/combine_drebin"
save_folder = "/scratch_NOT_BACKED_UP/NOT_BACKED_UP/xinran/dataset/drebin"
years = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

# Step 1:  df  sha256 -> vt_scan_date 
df['sha256'] = df['sha256'].astype(str)
df = df[['sha256', 'vt_scan_date']].drop_duplicates()

# 
sha2vt = dict(zip(df['sha256'], df['vt_scan_date']))

# Step 2:  vt_scan_date
for year in years:
    for month in months:
        file_path = os.path.join(dataset_folder, f'{year}-{month}', 'features.pkl')
        save_dir = os.path.join(save_folder, f'{year}-{month}')
        save_path = os.path.join(save_dir, 'features.pkl')

        if not os.path.exists(file_path):
            print(f"{file_path} not found, skipping...")
            continue

        with open(file_path, 'rb') as f:
            data = pickle.load(f)

        #  sha256 
        data['sha256'] = data['sha256'].astype(str)

        #  merge 
        data['vt_scan_date'] = data['sha256'].map(sha2vt)

        # 
        os.makedirs(save_dir, exist_ok=True)
        data.to_pickle(save_path)

        print(f"{year}-{month} done, columns: {list(data.columns)}")
```

```python
df_train['sha256'] = df_train['sha256'].astype(str)
df_test['sha256'] = df_test['sha256'].astype(str)

shared_sha256_train = set(df_train['sha256']) & set(df['sha256'])
shared_sha256_test = set(df_test['sha256']) & set(df['sha256'])
print(len(shared_sha256_train))
print(len(shared_sha256_test))
```

```python
df_shared_train = df[df['sha256'].isin(shared_sha256_train)].copy()
df_shared_test = df[df['sha256'].isin(shared_sha256_test)].copy()
df_vt_train = df_shared_train[['sha256', 'vt_scan_date']].drop_duplicates()
df_vt_test = df_shared_test[['sha256', 'vt_scan_date']].drop_duplicates()
df_train_merged = df_train.merge(df_vt_train, on='sha256', how='left')
df_test_merged = df_test.merge(df_vt_test, on='sha256', how='left')


print(df_train_merged.head())
print(df_test_merged.head())
```

```python
df_train_merged_malware = df_train_merged[df_train_merged['label'] == 1]
print(df_train_merged_malware.shape)
df_test_merged_malware = df_test_merged[df_test_merged['label'] == 1]
print(df_test_merged_malware.shape)
```
